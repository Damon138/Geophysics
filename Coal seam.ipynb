## Determining thickness and top depth of Coal Seam
Will be identifying thin coal beds using well log data based. The identification is based on Gamma ray response which is quite low, Bulk density response ranging from 1.1 gm/cc to 1.8 gm/cc and resistivity response which is quite large in case of coal.

```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio

#Hide warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
from tkinter import *
from tkinter import filedialog

# defining a function to select las file from your pc
def openfile():
    filepath=filedialog.askopenfilename(title="Select a file",
                                        filetypes=(('las files','*las'),
                                                   ('all','*.*')))
    return filepath  

root=Tk()
root.title('File Browser')
root.resizable(False, False)
root.geometry('300x150')
button=Button(text='Open',command=openfile)
button.pack()

file=openfile()
print(file)
root.mainloop()
```

```python
las_20=lasio.read(file)
```

```python
## Header information
for item in las_20.well:
    print(f"{item.descr} ({item.mnemonic}): {item.value}")
```

```python
well=las_20.df()
well
```

```python
well.columns
```

```python
well.reset_index(inplace=True)
```

```python
well.describe()
```

```python
def plot(top_depth,bottom_depth):
    logs = well[(well.DEPT >= top_depth) & (well.DEPT <= bottom_depth)]
    fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(18,23),sharey=True,gridspec_kw={'width_ratios':[1,1,1]})
    fig.suptitle("Well",fontsize=24)
    fig.subplots_adjust(top=0.9,wspace=0.1)

    # setting for all axs
    
    for axes in ax:
        axes.set_ylim(top_depth,bottom_depth)
        axes.invert_yaxis()
        axes.grid(True)
        axes.get_xaxis().set_visible(False)
        
        ax1=ax[0].twiny()
        ax1.set_xlim(0,200)
        ax1.spines['top'].set_position(('outward',0))
        ax1.plot(logs.SGRC,logs.DEPT,'-g')
        ax1.set_xlabel('SGRC (API)',color='g')
        ax1.tick_params(axis='x',colors='g')
        ax1.grid(True)
        
        ax11=ax[1].twiny()
        ax11.spines['top'].set_position(('outward',0))
        ax11.plot(logs.ALCDLQ,logs.DEPT,'-b')
        ax11.set_xlabel('ALCDLQ (g/cc)',color='b')
        ax11.tick_params(axis='x',colors='b')
        ax11.grid(True)
        
        ax12=ax[2].twiny()
        ax12.spines['top'].set_position(('outward',0))
        ax12.plot(logs.SEDA,logs.DEPT,'r')
        ax12.set_xlabel('SEDA (ohm-m)',color='r')
        ax12.set_xscale('log')
        ax12.tick_params(axis='x',colors='r')
        ax12.grid(True)
        
```

```python
plot(950,1600)
```

##### Identifing zone that satisfy a certain cutoff (condition)

```python
con = (well['SGRC']<22) * (well['ALCDLQ']<1.5) * ((well['SEDA']>150) * (well['SEDA']<200))
aa=[]
for i in range(0,4809):
    if con[i]==True:
        aa.append(well['DEPT'][i])
        
aa
```

```python
plot(1200,1400)
```

```python
bb=[]
for i in range(0,4810):
    if con[i]==True:
        bb.append(1)
    else:
        bb.append(0)
well['CLAS']=bb        
```

```python
plt.figure(figsize=(4,8))
plt.plot(well.CLAS,well.DEPT,'g')
# plt.xlim(0,1)
plt.ylim(1348,1340)
plt.fill_between(well.CLAS,well.DEPT,color='g',alpha=0.2)
plt.xlabel('COAL')
plt.ylabel('Depth (m)')
plt.grid(which='both')
```

##### Calculating top depth.

```python
top_depth=[]
for i in range(0,len(well.CLAS)-1):
    if well.CLAS[i+1]-well.CLAS[i]==1:
        top_depth.append(well.DEPT[i+1])
print(top_depth)
```

##### Calculating thickness of each zone.

```python
m = con
blocks = (~m).cumsum()[m]
blocks # creates a group of idndex with same value
```

```python
grp=well['DEPT'].groupby(blocks)
grp.groups
```

```python
data = list(grp.groups.items())
data
```

```python
# count 
for i in range(0,len(data)):
    print(len(data[i][1]))
```

```python
sample=well['DEPT'][10]-well['DEPT'][9]
sample
```

```python
# thickness of each zone
thickness=[]
for i in range(0,len(data)):
    thickness.append(len(data[i][1])*sample)
```

```python
pd.DataFrame(list(zip(top_depth,thickness)),columns=['Top_Depth','Thickness'])
```
